### 1. Create a line graph showing the development in poverty over time in both Argentina, Ivory cost, USA. (be aware of the missing data in particular years)

```python
%pylab inline
import pandas as pd
import subprocess

data = pd.read_csv('API_SI.POV.DDAY_DS2_en_csv_v2_10474275.csv',skiprows=3)
c_codes = data[data['Country Code'].isin(['ARG', 'CIV', 'USA'])]['Country Code']
graph = data.iloc[c_codes.index, 4:-1].T
graph = graph.rename(columns=dict(c_codes))
graph.plot()
```

### 2. Find the 10 countries with the highest poverty rate, and plot the rates in a bar chart

```python
year='2016'
poorest = data.sort_values(by=year,ascending=False).iloc[0:10][year].T
poorest.plot.bar()
```
